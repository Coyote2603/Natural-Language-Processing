## Stock Sentiment Analysis using News Headlines

```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/nehar/Downloads/Data (2).csv", encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
#Checking if it is imbalanced or balanced 
label_counts = df['Label'].value_counts()
print(label_counts)
```

```python
#time series dataset type
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
#DATA PREPROCESSING
```

```python
# Removing punctuations
#Taking dataset of all the features

data=train.iloc[:,2:27]

#apart from a-z replacing all other characters
#replaced all the special characters with blanks
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)


# Renaming column names for ease of access
#naming them 0,1,2....
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)

```

```python
# Convertng headlines to lower case
#same words can be detected 
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
#combining top 25 headlines in one row and combine them
#iterating through first sentence(row) and read all the column and use a join operation to join them with space in between them 
#' ' -space over there just to differentiate
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
#same for all the sentences (record)
#list headlines to store the data
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
#countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implementing BOW
#(2,2) COMBINING TWO FEATURE -ex-(girl boy )
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset[0]
#sparse matrix -in terms of 0,1
```

NOTE: Preprocessing has been done for only the train data but not the test data




```python
# implement RandomForest Classifier
#can perform hyperparameter tuning
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
#same program for the test dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
bowpredictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print("Accuracy score-",score)
report=classification_report(test['Label'],predictions)
print(report)
```

TF-IDF

```python
#TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implementing TFIDF
#(2,2) COMBINING TWO FEATURE -ex-(girl boy )
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
traindataset=tfidfvector.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
#can perform hyperparameter tuning
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
#same program for the test dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = tfidfvector.transform(test_transform)
tfidfpredictions = randomclassifier.predict(test_dataset)
```

```python
#predictions
predictions
```

```python
tfidfmatrix=confusion_matrix(test['Label'],predictions)
print(tfidfmatrix)
score=accuracy_score(test['Label'],tfidfpredictions)
print("Accuracy score of tfidf-",score)
tfidfreport=classification_report(test['Label'],tfidfpredictions)
print(tfidfreport)
```
